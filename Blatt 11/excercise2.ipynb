```python
def P1(t,n):
    sum = 0
    for k in range(-n,n+1,1):
        if k == 0:
            continue
        c = 2*pi*(1j/(pi*k))*(1-(-1)**k)
        sum += c*np.exp(-1j*k*t)
    return np.real(sum)

def P2(t,n):
    sum = 0
    for k in range(-n,n+1,1):
        if k == 0:
            continue
        c = (1/(pi*k**2))*((-1)**k-1)
        sum += c*np.exp(-1j*k*t)
    return np.real(sum)

x = np.linspace(-20,20,1000)

np.size(P2(x,5))
```

```python
fig, ax = plt.subplots(figsize=(15,5))

for i in range(1,11,1):
    x = np.linspace(0,30,1000)
    ax.plot(x,P1(x,i),label=i)

# x = np.linspace(0,30,1000)
# ax.plot(x,P1(x,100),label=i)

#plt.ylim(-1,4)
plt.title("$f_{1}(t)$ als Fourierreihe",fontsize=20)
ax.legend() 
ax.grid()
```
