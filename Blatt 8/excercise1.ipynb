```python
a = np.asfarray([-61.688, 72.5235, 72.822, -108.519, -5.12949, 39.9139, -7.07373, -3.91823, 1.0])
x = np.asfarray([0.9, -1.1,  1.4, 1.4, -2.0, -2.0, 2.2, 2.2])
```

```python
def P(x,a):
    sum = 0
    for i in range(len(a)):
        sum += a[i]*(x**i)
    return sum

def P_k(x,a,k,x0):
    sum = 1
    for i in range(k):
        sum = sum*(x-x0[i])
    sum = P(x,a)/sum
    return sum

def Pstrich(x,a):
    sum = 0
    for i in range(len(a)):
        if i-1 < 0:
            sum += 0
        else:
            sum += a[i]*i*(x**(i-1))
    return sum
```
